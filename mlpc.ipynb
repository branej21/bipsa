```python
directory="/kaggle/input/food-nutrition-dataset/FINAL FOOD DATASET/"
```

```python
import pandas as pd
import os



# List to hold the dataframes
dataframes = []

# Iterate over all the files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory, filename)
        # Read the CSV file and append to the list
        df = pd.read_csv(file_path)
        
        dataframes.append(df)

# Concatenate all dataframes in the list
combined_df = pd.concat(dataframes, ignore_index=True)


```

## Correlation Matrix

```python

# Calculate the correlation matrix excluding non-numeric columns
correlation_matrix = combined_df[numeric_columns].corr()

# Plot the heatmap
plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### Hishest Correlation

```python
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Apply the mask to the correlation matrix
corr_matrix_masked = correlation_matrix.mask(mask)

# Find the highest correlated pairs
highest_corr_pairs = corr_matrix_masked.unstack().dropna().sort_values(ascending=False).reset_index()
```

```python
highest_corr_pairs.columns = ['Feature 1', 'Feature 2', 'Correlation']
highest_corr_pairs.head(10)
```

```python
most_correlated_features = highest_corr_pairs.head(10)[['Feature 1', 'Feature 2']].values.flatten()

# Keep only the relevant columns for the most correlated features
corr_subset = correlation_matrix.loc[most_correlated_features, most_correlated_features]

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_subset, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Most Correlated Components Matrix')
plt.show()
```

## Boxplots

```python
# Boxplots of numeric columns
plt.figure(figsize=(20, 15))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(6, 6, i)
    sns.boxplot(y=combined_df[col])
    plt.title(col)
plt.tight_layout()
plt.suptitle('Boxplots of Numeric Features', y=1.02)
plt.show()
```

## Nutritional Density Analysis

```python
food_groups = combined_df.groupby('food').mean().reset_index()
```

```python
# Top food items by nutritional density
top_nutrition_density_foods = food_groups.sort_values(by='Nutrition Density', ascending=False).head(10)
plt.figure(figsize=(12, 8))
sns.barplot(x='Nutrition Density', y='food', data=top_nutrition_density_foods, palette='viridis')
plt.title('Top 10 Food Items by Nutrition Density')
plt.show()
```
